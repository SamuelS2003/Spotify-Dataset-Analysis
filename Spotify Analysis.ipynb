```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df
```

```python
#Genre Analysis

genre_popular = df.groupby('genre')['popularity'].mean().head(10).sort_values(ascending=False)

#Plotting
plt.figure(figsize=(10,7))

plt.bar(genre_popular.index, genre_popular.values, color ='blue')
plt.title('Genre Popularity')
plt.xlabel('Genre')
plt.ylabel('Popularity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
#Cleaning the Artist Coloumn
df['artists'] = df['artists'].str.split(', ')
df = df.explode('artists').reset_index(drop=True)

#Artist Popularity Analysis

artist_popular = df.groupby('artists')['popularity'].mean().head(10).sort_values(ascending=False)

#Plotting
plt.figure(figsize=(10,7))

plt.bar(artist_popular.index, artist_popular.values, color = ['red','tomato'])
plt.title('Artist Popularity')
plt.xlabel('Artists')
plt.ylabel('Popularity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
#Album Popularity Analysis

album_popular = df.groupby('album')['popularity'].mean().head(10).sort_values(ascending=False)

#Plotting
plt.figure(figsize=(7,7))

plt.bar(album_popular.index, album_popular.values, color = ['red','tomato'])
plt.title('Album Popularity', fontsize=15)
plt.xlabel('Album')
plt.ylabel('Popularity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
#Album Playtime Analysis

album_playtime = df.groupby('album')['duration_ms'].mean().head(10).sort_values(ascending=False)

#Plotting
plt.figure(figsize=(7,7))

plt.bar(album_playtime.index, album_playtime.values, color = ['Blue','tomato'])
plt.title('Album Playtime', fontsize=15)
plt.xlabel('Album')
plt.ylabel('Playtime in Milliseconds')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
#Artist and Album Analysis
most_album = df.groupby('artists')['album'].nunique().sort_values(ascending=False).head(10)
sns.barplot(x=most_album.values, y=most_album.index, palette='viridis')
plt.title('Artist with most albums')
plt.ylabel('Artists')
plt.show()
```

```python
#Explicit Content Analysis
explicit = df['explicit'].value_counts()
plt.pie(
    explicit, 
    labels=explicit.index, 
    autopct='%1.f%%', 
    colors=['Red','tomato']
)
plt.title('Explicit Content Ratio')
plt.show()
```

```python
#Albums with Explict Content
explicit_albums = df[df['explicit']==True]

artist_explicit = explicit_albums.groupby('album')['popularity'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,7))
sns.barplot(x=artist_explicit.values, y=artist_explicit.index, palette = 'mako')
plt.title('Explicit Albums Ranking', fontsize=15)
plt.ylabel('Albums', fontsize = 15)
plt.show()
```
